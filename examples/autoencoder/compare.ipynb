# Code to compare the results of two notebooks
In this notebook we compare the results between "train_baseline.ipynb" and "train_RRAE.ipynb" to show the effect of the SVD

We compare the capabilities of both autoencoders in both interpolation and random generation in the latent space

```python
import matplotlib.pyplot as plt
```

```python
# Interpolation
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
img = plt.imread("interpolation_no_SVD.png")
plt.imshow(img)
plt.axis('off')
plt.title("Baseline Autoencoder")
plt.subplot(1, 2, 2)
img = plt.imread("interpolation_RRAE.png")
plt.imshow(img)
plt.axis('off')
plt.title("RRAE Autoencoder")
plt.show()
```

```python
# Random generation
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
img = plt.imread("random_sample_no_SVD.png")
plt.imshow(img)
plt.axis('off')
plt.title("Baseline Autoencoder")
plt.subplot(1, 2, 2)
img = plt.imread("random_sample_RRAE.png")
plt.imshow(img)
plt.axis('off')
plt.title("RRAE Autoencoder")
plt.show()
```

# Discussion
While neither Autoencoders are perfect, we can clearly see that the RRAE has a better latent space.

- For interpolation, the baseline autoencoder splits the gaussians in two while this doesn't happen in RRAEs.

- For random generation, while the gaussians generated by RRAEs are a bit noisy, they are of much better quality than the ones generated by the baseline Autoencoder

## The reasons

The SVD in this case has two main roles:

- It creates a bottleneck of dimension 2 (the rank chosen), which did not exist in the baseline

- It regularizes the latent space since the compressed representation of the SVD has a special form.
